```python
import pandas as pd
import json

# LOAD CSV
df = pd.read_csv("chatbot_data.csv")

# ONLY REQUIRED COLUMNS
df = df[['message', 'intent']]

intents = []

# GROUP BY INTENT
grouped = df.groupby('intent')

for intent, group in grouped:

    patterns = group['message'].tolist()

    intent_data = {
        "tag": intent,

        "patterns": patterns,

        "responses": [
            f"This is response for {intent}"
        ]
    }

    intents.append(intent_data)

# FINAL JSON
data = {
    "intents": intents
}

# SAVE JSON
with open("responses.json", "w") as f:

    json.dump(data, f, indent=4)

print("responses.json created successfully")
```
